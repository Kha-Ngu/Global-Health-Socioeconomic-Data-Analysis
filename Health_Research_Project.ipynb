**<h3> Below is the analysis of he correlation between Income Composition and Healthcare Expenditure by Countries**

```python
import seaborn as sns

# Income Composition vs. Total Health Expenditure
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sdf2, x='Income composition of resources', y='Total expenditure', hue='Status', palette='Spectral')
plt.title('Income Composition vs. Total Health Expenditure')
plt.xlabel('Income Composition of Resources')
plt.ylabel('Total Health Expenditure (% of GDP)')
plt.legend(title='Status')
plt.grid(True)
plt.show()
corr_val = sdf2['Income composition of resources'].corr(sdf2['Total expenditure'])
print('R-value: ' + str(corr_val))
sns.lmplot(x="Income composition of resources", y="Total expenditure", hue='Status', data=sdf2);
print('Below shows the line of best fit')
```

**Analysis**
<br> With such a low correlation value, it seems that there is almost no correlation between countries with high income earners and how much the country spends on its health industry. This suggests that healthcare access is still an individual issue at large, and goverment expenditure have almost no effect on effect on the life longevity of its citizens.

**<h3>Observable Patterns**
<br><p>A clear trend emerges when we compare the graphs: <br>
1. Developed countries such as Australia and Germany have extremely low infection rates of HIV/AIDS and extremely high immunization rates for Hepatitis B.
2. Developed countries have higher income composition, and their total health expenditure are also higher as a result.
3. While the immunization rates for Hepatitis B between countries are somewhat similar, the starkest difference is in the infection rates of HIV/AIDS between developing and developed countries.
4. Developed countries both have high income composition and health expenditures, but it is clear that those two variables are not associative.
5. Despite the massive difference in some developed countries' GDPs, they seem to spend around the same percentage of their GDP on health indsutries.

**<h3>1. Income Composition vs. Life Expectancy**

```python
# Income Composition vs. Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sdf2, x='Income composition of resources', y='Life expectancy ', hue='Status', palette='Spectral')
plt.title('Income Composition vs. Life Expectancy')
plt.xlabel('Income Composition of Resources')
plt.ylabel('Life Expectancy (in years)')
plt.legend(title='Status')
plt.grid(True)
plt.show()
corr_val = sdf2['Income composition of resources'].corr(sdf2['Life expectancy '])
print('R-value: ' + str(corr_val))
sns.lmplot(x="Income composition of resources", y="Life expectancy ", hue='Status', data=sdf2);
print('Below shows the line of best fit')
```

##**Analysis**
<p>The scatter plot illustrates the relationship between income composition of resources and life expectancy across different countries. The largest group observed in the plot comprises developing countries, which tend to have lower income composition of resouces and shorter life expectancy compared to developed countries. A positive correlation is evident with the r-value being 0.89, indicating a strong positive correlation. This suggests that countries with higher income composition of resouces generally exhibit higher life expectancy and shows that individual economic prosperity is associated with better health outcomes and longevity. <br><br>
This trend seems to be the same for both developing and developed countries, however developing countries seem to have a steeper slope for its line of best fit, and the data are less scattered around the line than the data in the developed countries, which indicates that citizens in developing countries are more dependent on their income for healthcare access than those living in developed countries. Thus, a person's socioeconomic standing in developing countries are more vital to their life expectancy than those in developed countries, which might have access to better health infrastructure.
</p>

**<h3>2. Healthcare Expenditure vs. Life Expectancy**

```python
# Scatter plot of Healthcare Expenditure vs. Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sdf2, x='Total expenditure', y='Life expectancy ', hue='Status', palette='coolwarm')
plt.title('Healthcare Expenditure vs. Life Expectancy')
plt.xlabel('Total Healthcare Expenditure (% of GDP)')
plt.ylabel('Life Expectancy (in years)')
plt.legend(title='Status')
plt.grid(True)
plt.show()
corr_val = sdf2['Total expenditure'].corr(sdf2['Life expectancy '])
print('R-value: ' + str(corr_val))
sns.lmplot(x="Total expenditure", y="Life expectancy ", hue='Status', data=sdf2);
print('Below shows the line of best fit:')
```

##**Analysis**

There is a positive correlation observed, however the correlation is weak with an r-value of 0.32. This indicates that higher investment in healthcare is somewhat associated with longer life expectancy, however it is not the main factor for improving life longevity.<br> <br>
This is especially true with developing countries, as their data points are immensely scattered around the line of best fit, suggesting that investing money into healthcare industries alone is not enough in developing countries. For developed countries, which adheres to the line of best fit better, it seems that health expenditure does play a positive role when other factors as other health-affecting life factors are most likely less severe in these countries.</p>
